```python
                                                                        #importing libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori                                              #apyori library
```

```python
store_data = pd.read_excel('train.xlsx')                                #creating data frame of training data
```

```python
store_data.head()
```

```python
store_data = pd.read_excel('train.xlsx', header=None)                     #removing header
```

```python
store_data.head()
```

```python
records = []                                                              #preprocessing of data into List
for i in range(1,15):
    records.append([str(store_data.values[i,j]) for j in [1,3,9,10,-1]])
```

```python
records
```

['Age', 'Department', 'JobSatisfaction', 'MaritalStatus', 'Attrition']

```python
                                                                          #genearating association rules
association_rules = apriori(records, min_support=0.07, min_confidence=0.7, min_lift=3, min_length=3)
association_results = list(association_rules)
```

```python
association_results
```

```python
print(association_results[0])                                             #printing first association rule
```

```python
for item in association_results:                                          #printing association rules
    pair = item[0] 
    items = [x for x in pair]
 
    if(len(items)==2):
        print (items[0] + '->' + items[1])
    if(len(items)==3):
        print (items[0] + '->' + items[1] + '->' + items[2])
    if(len(items)==4):
        print (items[0] + '->' + items[1] + '->' + items[2] + '->' + items[3])
    if(len(items)>4):
        print (items[0] + '->' + items[1] + '->' + items[2] + '->' + items[3] + '->' + items[4])
        
    print("Support: " + str(item[1]))
    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```
